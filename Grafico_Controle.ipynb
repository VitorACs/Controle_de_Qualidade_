```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
Dados_df = pd.read_excel('DataSet_Grafico_Controle.xlsx',sheet_name='Exercicio 1')
display(Dados_df)
```

```python
Colunas = np.array(Dados_df.columns)
Colunas_retiradas = np.delete(Colunas,[3,15,24])
 
Amostras = []
for coluna in Colunas:
    Amostras.append([coluna,np.array(Dados_df[coluna])])
    

print(len(Colunas_retiradas))
```

### Calcular limites

```python
Amplitude = [np.max(Amostras[i][1]) - np.min(Amostras[i][1]) for i in range(len(Amostras))]
Amplitude_ = np.mean(Amplitude)

Media = [np.mean(Amostras[i][1]) for i in range(len(Amostras))]
Media_ = np.mean(Media)

# esses valores estão para uma amostra de tamanho n
LSC_amplitude = 2.115 * Amplitude_
LIC_amplitude = 0 * Amplitude_

# esses valores estão para uma amostra de tamanho n
LSC_media = Media_ + 0.577 * Amplitude_
LIC_media = Media_ - 0.577 * Amplitude_

print(Amplitude_)
```

### Plotar o gráfico

```python
# gráfico amplitude
fator_correção = 0.1


mosaico = "AB"
fig = plt.figure(figsize=(16, 5))
axs = fig.subplot_mosaic(mosaico)

axs['A'].plot(range(len(Amplitude)),Amplitude,marker='o',color='Darkgreen')

#Linha da média
axs['A'].axhline(Amplitude_,0,color='black')
axs['A'].annotate(f'A = {Amplitude_:.2f}',xy=(np.max(range(len(Amplitude)))+1,Amplitude_+fator_correção))
#Linha superior
axs['A'].axhline(LSC_amplitude,0,color='grey',linestyle='--')
axs['A'].annotate(f'LSC = {LSC_amplitude:.2f}',xy=(np.max(range(len(Amplitude))),LSC_amplitude+fator_correção))
#Linha inferior
axs['A'].axhline(LIC_amplitude,0,color='grey',linestyle='--')
axs['A'].annotate(f'LIC = {LIC_amplitude:.2f}',xy=(np.max(range(len(Amplitude))),LIC_amplitude+fator_correção))

axs['A'].set_title('Gráfico de Amplitude', size=14)
axs['A'].set_xlabel('N° de amostras')
axs['A'].set_ylabel('Amplitude')


axs['A'].set_xlim(-1,len(Amplitude)+4)
axs['A'].set_ylim(LIC_amplitude-0.8,LSC_amplitude+0.8)

#----------------------------------------------------------------------------------------------------------------------------------------


axs['B'].plot(range(len(Media)),Media,marker='o',color='Darkblue')

#Linha da média
axs['B'].axhline(Media_,0,color='black')
axs['B'].annotate(f'M = {Media_:.2f}',xy=(np.max(range(len(Media)))+1,Media_+fator_correção))
#Linha superior
axs['B'].axhline(LSC_media,0,color='grey',linestyle='--')
axs['B'].annotate(f'LSC = {LSC_media:.2f}',xy=(np.max(range(len(Media)))+0.5,LSC_media+fator_correção))
#Linha inferior
axs['B'].axhline(LIC_media,0,color='grey',linestyle='--')
axs['B'].annotate(f'LIC = {LIC_media:.2f}',xy=(np.max(range(len(Media))),LIC_media+fator_correção))

axs['B'].set_title('Gráfico da Média', size=14)
axs['B'].set_xlabel('N° de amostras')
axs['B'].set_ylabel('Média')


axs['B'].set_xlim(-1,len(Amplitude)+4)
axs['B'].set_ylim(LIC_media-0.8,LSC_media+0.8)


fig.suptitle(f'Peso na embalagem de aditivos \n\n\n\n',weight='bold',size = 15,)
plt.show()

fig.savefig('Exercicio1.png',dpi=300,bbox_inches='tight')
```

```python
Exercicio_2 = pd.read_excel('DataSet_Grafico_Controle.xlsx',sheet_name='Exercicio 2')
display(Exercicio_2)
```

```python

m = np.array(Exercicio_2['Amostra'])
n = np.array(Exercicio_2['Tamanho da amostra'])
d = np.array(Exercicio_2['Latas Fora do Padrão'])
p = [d[i]/n[i] * 50 for i in range(len(n))] 
p_ = np.sum(d)/(np.max(m)*np.max(n))
```

```python
LMS = lambda x,y: x*y + 3*np.sqrt((x*y*(1-y)))

def LMI(x,y):
    LMI = x*y - 3*np.sqrt((x*y*(1-y)))
    if LMI <= 0:
        return 0
    else:
        return LMI

P_ = np.max(n)*p_
print(LMS(np.max(n),p_))
```

```python
plt.figure(figsize=(10,5))
plt.plot(m,p,marker='o',color='green')

plt.axhline(p_*np.max(n),0,np.max(m),color='black')
plt.annotate(f'M = {p_*np.max(n)}',(np.max(m)+1,p_*50+0.5))

plt.axhline(LMS(np.max(n),p_),0,np.max(m),color='grey',linestyle='--')
plt.annotate(f'LMS = {LMS(np.max(n),p_):.2f}',(np.max(m),LMS(np.max(n),p_)+0.5))
plt.axhline(LMI(np.max(n),p_),0,np.max(m),color='grey',linestyle='--')
plt.annotate(f'LMI = {LMI(np.max(n),p_):.2f}',(np.max(m),LMI(np.max(n),p_)+0.5))

plt.title('Gráfico p', weight='bold', size=14)
plt.xlabel('N° de Amostras')
plt.ylabel('Proporção Amostral fora do Padrão')

plt.xlim(-1,np.max(m)+4.5)
plt.ylim(-1,15)

plt.savefig('Exercicio2.png',dpi=300,bbox_inches='tight')
```

```python
LMS_atualizado = (LMS(np.max(n),p_) - p_*np.max(n))/2 + p_*np.max(n)
LMI_atualizado = (p_*np.max(n)-LMS(np.max(n),p_))/2 + p_*np.max(n)
print(LMI_atualizado,p_)
```

```python
plt.figure(figsize=(10,5))
plt.plot(m,p,marker='o',color='green')

plt.axhline(p_*np.max(n),0,np.max(m),color='black')
plt.annotate(f'M = {p_*np.max(n)}',(np.max(m)+1,p_*50+0.5))

plt.axhline(LMS_atualizado,0,np.max(m),color='grey',linestyle='--')
plt.annotate(f'LMS = {LMS_atualizado:.2f}',(np.max(m),LMS_atualizado+0.5))
plt.axhline(LMI_atualizado,0,np.max(m),color='grey',linestyle='--')
plt.annotate(f'LMI = {LMI_atualizado:.2f}',(np.max(m),LMI_atualizado+0.5))

plt.title('Gráfico p', weight='bold', size=14)
plt.xlabel('N° de Amostras')
plt.ylabel('Proporção Amostral fora do Padrão')

plt.xlim(-1,np.max(m)+4.5)
plt.ylim(-1,15)

plt.savefig('Exercicio2-1.png',dpi=300,bbox_inches='tight')
```
